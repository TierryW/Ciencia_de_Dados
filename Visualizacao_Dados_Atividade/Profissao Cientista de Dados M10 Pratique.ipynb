```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('ecommerce_preparados.csv')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.width', None)

print(df.head(20).to_string())
```

```python
df = df.dropna()
print(df.head(10).to_string())
print(df.isnull().sum())
```

```python
# Gráfico de Disersão
plt.scatter(df['Nota_MinMax'], df['N_Avaliações_MinMax'], color = "#510073", alpha = 0.6, s=50)
plt.title('Gráfico Dispersão - Nota e N_Avaliações')
plt.xlabel('Nota_MinMax')
plt.ylabel('N_Avaliações_MinMax')
plt.show()
```

```python
# Mapa de Calor
corr = df[['N_Avaliações', 'Qtd_Vendidos_Cod']].corr()
sns.heatmap(corr,annot=True, cmap='GnBu')
plt.title('Correlação N_Avaliações e Qtd_vendidos')
plt.tight_layout()
plt.show()
```

```python
# Gráfico de barras
plt.figure(figsize=(10,6))
df['Nota'].value_counts().plot(kind='bar', color="#f3ef80")
plt.title('Quantidade de Avaliações')
plt.xlabel('Notas')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.show()
```

```python
df['Temporada'] = df['Temporada'].apply(lambda x: x if x in ['primavera/verão', 'outono/inverno'] else 'não definido')
x = df['Temporada'].value_counts().index
y = df['Temporada'].value_counts().values
```

```python
# Gráfico pizza
plt.figure(figsize=(7, 6))
plt.pie(y, labels=x, autopct='%1.1f%%', startangle=90)
plt.title('Temporada')
plt.show()
```

```python
# Gráfico de Densidade
plt.figure(figsize=(10,6))
sns.kdeplot(df['Nota'], fill=True, color="#f2bd5a")
plt.title('Densidade das Notas')
plt.xlabel('Notas')
plt.show()
```
