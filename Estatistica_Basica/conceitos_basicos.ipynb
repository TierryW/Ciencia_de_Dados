```python
import pandas as pd
```

```python
df = pd.read_csv('ESPORTISTAS_MOD7.csv', delimiter=';')
df.head()
```

```python
df.groupby('Esporte')['Atleta'].count().reset_index().sort_values(by='Atleta', ascending=False)
```

```python
contagem = df['Esporte'].value_counts(normalize=True).reset_index()
contagem
```

## Média

```python
soma_ganhos_totais = df['Ganhos_Totais'].sum()
quantidade_registros = len(df['Ganhos_Totais'])
media_ganhos_totais = soma_ganhos_totais / quantidade_registros
print('Média de Ganhos Totais: ', media_ganhos_totais)
```

```python
df['Ganhos_Totais'].mean()
```

```python
df.groupby('Esporte')['Ganhos_Totais'].mean().reset_index().sort_values(by='Ganhos_Totais', ascending=False)
```

## Mediana

```python
df['Ganhos_Totais'].median()
```

```python
df.groupby('Esporte')['Ganhos_Totais'].median().reset_index().sort_values(by='Ganhos_Totais', ascending=False)
```

## Moda

```python
df['Ganhos_Totais'].mode()
```

## Desvio Padrão

```python
desvio_padrao = df.groupby('Esporte')['Ganhos_Totais'].std().reset_index()
desvio_padrao
```

### Alguns registros estão NAN, isso ocorre quando temos apenas 1 registro para o esporte, pois não temos como calcular o desvio padrão se não temos mais de um registro
