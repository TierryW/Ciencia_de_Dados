## Moda

```python
calcular_moda <- function(x){
  unique_x <- unique(x)  # unique() para obter os valores únicos do vetor.
  contagem <- tabulate(match(x, unique_x))  # tabulate() para cotar a frequencia de cada valor.
  moda <- unique_x[which.max(contagem)]  # which.max() para encontrar o índice do valor com a frequencia máxima.
  return(moda)
}
calcular_moda(dados$Ganhos_Totais)
```
