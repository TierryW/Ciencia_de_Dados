```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## DataFrame

```python
idades = np.random.randint(20, 60, 20)
salarios = idades * np.random.randint(80, 120, 20)
pontuacoes = salarios * np.random.uniform(0.5, 1.5, 20)
profissoes = ['Engenheiro', 'Professor', 'Médico', 'Advogado', 'Designer', 'Analista', 'Gerente', 'Programador']

novo_data = {'Idade': idades, 'Salário': salarios, 'Pontuação': pontuacoes.round(2)}  # Arrendondando as pontuações para 2 casas decimais
df = pd.DataFrame(novo_data)
df['Profissão'] = np.random.choice(profissoes, size=len(df))
df.head(10)
```

## Gráfico de Barras

```python
plt.figure(figsize=(10, 6))
salario_profissao = df.groupby('Profissão')['Salário'].mean()
plt.bar(salario_profissao.index, salario_profissao, color='skyblue')
plt.title('Distribuição de Salários', fontsize=16)
plt.xlabel('Índice', fontsize=12)
plt.ylabel('Salário', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)  # Adicionando grades ao gráfico com estilo de linha tracejada (--)
plt.show()
```

## Gráfico de Linhas

```python
media_score_idade = df.groupby('Idade')['Pontuação'].mean()
media_score_idade = media_score_idade.sort_index()

plt.figure(figsize=(10, 6))
plt.plot(media_score_idade.index, media_score_idade, marker='o', color='red')
plt.title('Média de Score por Idade', fontsize=16)
plt.xlabel('Idade', fontsize=12)
plt.ylabel('Média do Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## Gráfico de Linhas e Colunas

```python
media_salario_idade = df.groupby('Idade')['Salário'].mean()
media_salario_idade = media_salario_idade.sort_index()

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(media_salario_idade.index, media_salario_idade, marker='o', color='red', label='Média do Score')
ax1.set_xlabel('Idade', fontsize=12)
ax1.set_ylabel('Média do Score', color='red', fontsize=12)

ax2 = ax1.twinx()
ax2.bar(media_salario_idade.index, media_salario_idade, color='blue', alpha=0.5, label='Média do Salário')
ax2.set_ylabel('Média do Salário', color='blue', fontsize=12)

# Legendas
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')
plt.title('Média do Score e Média do Sálario por Idade', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(media_salario_idade.index, media_salario_idade, marker='o', color='red', label='Média do Score')
ax1.set_xlabel('Idade', fontsize=12)
ax1.set_ylabel('Média do Score', color='red', fontsize=12)

ax2 = ax1.twinx()
ax2.bar(media_salario_idade.index, media_salario_idade, color='blue', alpha=0.5, label='Média do Salário')
ax2.set_ylabel('Média do Salário', color='blue', fontsize=12)

# Legendas
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right', bbox_to_anchor=(1.2, 1.2))  # Especifica onde a legenda será posicionada
plt.title('Média do Score e Média do Sálario por Idade', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## Gráfico de Dispersão

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Idade'], df['Salário'], color='purple', alpha=0.7)

plt.title('Relação entre Salário e Idade', fontsize=16)
plt.xlabel('Idade', fontsize=12)
plt.ylabel('Salário', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## BoxPlot

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['Salário'])

plt.title('BoxPlot do Salário', fontsize=16)
plt.ylabel('Salário', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['Pontuação'])

plt.title('BoxPlot da Pontuação', fontsize=16)
plt.ylabel('Pontuação', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot([df['Pontuação'], df['Salário']], labels=['Pontuação', 'Salário'])

plt.title('BoxPlot da Pontuação e Salários', fontsize=16)
plt.ylabel('Valor', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Simulando um DataFrame com Outliers
salarios = np.random.randint(2000, 10000, 18)
outliers =  [15000, 18000]

salarios_outliers = np.concatenate((salarios, outliers))
df_salarios = pd.DataFrame({'Salário': salarios_outliers})
df_salarios.head(5)
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df_salarios['Salário'])

plt.title('BoxPlot dos Salários', fontsize=16)
plt.ylabel('Salários', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
