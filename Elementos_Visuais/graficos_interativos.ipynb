```python
import pandas as pd
import numpy as np
import plotly.express as px
```

## Uso da biblioteca plotly.express para gráficos interativos

## DataFrame

```python
idades = np.random.randint(20, 60, 20)
salarios = idades * np.random.randint(80, 120, 20)
pontuacoes = salarios * np.random.uniform(0.5, 1.5, 20)
profissoes = ['Engenheiro', 'Professor', 'Médico', 'Advogado', 'Designer', 'Analista', 'Gerente', 'Programador']
estado_civil = ['Casado', 'Solteiro']

novo_data = {'Idade': idades, 'Salário': salarios, 'Pontuação': pontuacoes.round(2)}  # Arrendondando as pontuações para 2 casas decimais
df = pd.DataFrame(novo_data)
df['Profissão'] = np.random.choice(profissoes, size=len(df))
df['Estado Civil'] = np.random.choice(estado_civil, size=len(df))
df.head(10)
```

## Gráfico de Barras na Horizontal

```python
salario_profissao = df.groupby('Profissão')['Salário'].mean().reset_index()

fig = px.bar(salario_profissao, x='Salário', y='Profissão', orientation='h', 
            title='Salários por Profissão',
            labels={'Salário': 'Salário Médio', 'Profissão': 'Profissão'})
fig.show()
```

## Gráficos de Barras na Horizontal com cores

```python
salario_profissao = df.groupby('Profissão')['Salário'].mean().reset_index()

fig = px.bar(salario_profissao, x='Salário', y='Profissão', orientation='h', 
            title='Salários por Profissão',
            labels={'Salário': 'Salário Médio', 'Profissão': 'Profissão'},
            color='Salário', width=800)  # Especificando a cor baseada nos valores de salário e definindo a largura do gráfico em pixels
fig.show()
```

## Gráfico Treemap

```python
salario_profissao_estado_civil = df.groupby(['Profissão', 'Estado Civil'])['Salário'].mean().reset_index()

# Gráfico de treemap com os salários ppor profissões usando as cores para representar o estado civil
fig = px.treemap(salario_profissao_estado_civil, path=['Profissão', 'Estado Civil'],
                values='Salário', title='Salários por Profissão e Estado Civil',
                color='Estado Civil')
fig.show()
```

## Gráfico BoxPlot

```python
fig = px.box(df, y='Idade', title='BoxPlot de Idade')
fig.show()
```
